```python
from gtts import gTTS
import os

# 1. Define project structure
project_dir = 'my_audio_project_single_code'
text_files_dir = 'text_files'
audio_output_dir = 'audio_output'

text_output_path = os.path.join(project_dir, text_files_dir)
audio_full_output_path = os.path.join(project_dir, audio_output_dir)

# Create directories
os.makedirs(text_output_path, exist_ok=True)
os.makedirs(audio_full_output_path, exist_ok=True)

print(f"Directory structure created under '{project_dir}':")
print(f"- {project_dir}/{text_files_dir}/")
print(f"- {project_dir}/{audio_output_dir}/\n")

# 2. Generate multiple text files
texts = [
    "This is the first sentence for audio, generated from a single code block.",
    "This is the second file content, now combined for simplicity.",
    "And this is the third text, converted to audio in one go.",
    "A fourth file for good measure, with a simple statement, all in one script."
]

print(f"Creating text files in: {text_output_path}/")
text_files_created = []
for i, text_content in enumerate(texts):
    file_name = f"text_file_{i+1}.txt"
    full_file_path = os.path.join(text_output_path, file_name)
    with open(full_file_path, 'w') as f:
        f.write(text_content)
    text_files_created.append(file_name)
    print(f"Created: {file_name}")

print(f"Text files created successfully in {text_output_path}/\n")

# 3. Convert Text Files to Audio
print(f"Converting text files from '{text_output_path}' to audio in '{audio_full_output_path}'...")

for text_file_name in text_files_created:
    full_text_file_path = os.path.join(text_output_path, text_file_name)

    # Read the content of the text file
    with open(full_text_file_path, 'r') as f:
        text_content = f.read()

    # Create a gTTS object
    tts = gTTS(text=text_content, lang='en')

    # Construct output filename for the audio file
    audio_file_name = text_file_name.replace('.txt', '.mp3')
    full_audio_file_path = os.path.join(audio_full_output_path, audio_file_name)

    # Save the generated audio
    tts.save(full_audio_file_path)
    print(f"Converted '{text_file_name}' to '{audio_file_name}'")

print("\nAll text files converted to audio successfully!")

# Verify the creation of the audio files
print(f"\nListing contents of '{audio_full_output_path}':")
!ls -1 "{audio_full_output_path}"

print(f"\nListing contents of '{text_output_path}':")
!ls -1 "{text_output_path}"
```

```python
from gtts import gTTS
import os

# Define the input folder (where your existing text files are)
input_folder = "my_audio_project_single_code/text_files"
# Define the output folder for the generated audio files
output_folder = "my_audio_project_single_code/audio_output"

# Ensure the output directory exists
os.makedirs(output_folder, exist_ok=True)

print(f"Processing text files from: '{input_folder}'")
print(f"Audio output will be saved to: '{output_folder}'\n")

# Check if the input folder exists
if not os.path.exists(input_folder):
    print(f"Error: The input folder '{input_folder}' was not found. Please ensure it exists and contains your text files.")
elif not os.listdir(input_folder):
    print(f"Warning: The input folder '{input_folder}' is empty. No text files to convert.")
else:
    # Get a list of all .txt files in the input folder
    text_files = [f for f in os.listdir(input_folder) if f.endswith('.txt')]

    if not text_files:
        print(f"No .txt files found in '{input_folder}'.")
    else:
        print("Converting text files to audio:")
        for fname in text_files:
            file_path = os.path.join(input_folder, fname)
            with open(file_path, 'r') as f:
                text = f.read().strip()

            if len(text) == 0:
                print(f"Skipping empty file: {fname}")
                continue

            try:
                tts = gTTS(text=text, lang ="en")
                base_name = os.path.splitext(fname)[0]
                output_path = os.path.join(output_folder, base_name + ".mp3")

                tts.save(output_path)
                print("Created:", output_path)
            except Exception as e:
                print(f"Error creating audio for {fname}: {e}")

        print("\nAll processing complete!")
        print(f"\nListing generated audio files in '{output_folder}':")
        !ls -1 "{output_folder}"
```

```python
!pip install gtts
from gtts import gTTS
import os
input_text_file = "/content/textfiles.txt"
output_audio_dir = "audio_from_user_textfile"
os.makedirs(output_audio_dir, exist_ok=True)

print(f"Processing input file: '{input_text_file}'")
if not os.path.exists(input_text_file):
    print(f"Error: The input file '{input_text_file}' was not found. Please ensure it exists in the '/content/' directory.")
else:
    with open(input_text_file, 'r') as f:
        lines = f.readlines()
    if not lines:
        print(f"The file '{input_text_file}' is empty. No audio files will be generated.")
    else:
        for i, line in enumerate(lines):
            processed_line = line.strip()
            if processed_line:
                try:
                    tts = gTTS(text=processed_line, lang='en')
                    audio_filename = f"audio_line_{i+1}.mp3"
                    output_path = os.path.join(output_audio_dir, audio_filename)
                    tts.save(output_path)
                    print(f"Generated: {output_path} for line {i+1}")
                except Exception as e:
                    print(f"Error generating audio for line {i+1}: {e}")
            else:
                print(f"Skipping empty line {i+1} in '{input_text_file}'.")

    print("\nAll lines processed. Audio generation complete.")
    print(f"Listing generated audio files in '{output_audio_dir}':")
    !ls -1 "{output_audio_dir}"
```
